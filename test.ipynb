```python
import numpy as np
import idx2numpy
from predict import make_prediction
```

```python
test_images = idx2numpy.convert_from_file("data/testimages.idx3-ubyte")
test_labels = idx2numpy.convert_from_file("data/testlabels.idx1-ubyte")
```

```python
accuracy = 0
for i in range(test_images.shape[0]):
    if make_prediction(test_images[i]) == test_labels[i]:
        print("Correct!")
        accuracy += 1 / test_images.shape[0]
    else:
        print("Incorrect.")
print(f'Accuracy of this model on the test data is equal to {accuracy}')
```
